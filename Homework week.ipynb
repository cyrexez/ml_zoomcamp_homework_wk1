```python
select_asians=asian_cars_update[["vehicle_weight","model_year"]]
select_asians.head(7)
x=select_asians.head(7).values
print (x)
```

```python
import numpy as np
xtx = np.dot(x.T, x)
print(xtx)
```

```python
xtx_inverse = np.linalg.inv(xtx)
print(xtx_inverse)
```

```python
y = np.array([1100, 1300, 800, 900, 1000, 1100, 1200])
print(y)
```

```python
w = np.dot(np.dot(xtx_inverse,x.T),y)
```
